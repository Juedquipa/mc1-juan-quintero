# Regresión lineal

## Definir valores de x y y

```python
import numpy as np

x_values = np.array([1,2,3,4,5])
y_values = np.array([3.5, 5.5, 6.8, 7.8, 8.3])

if(len(x_values)!=len(y_values)):
    print('La longitud de los puntos de x y y no es la misma.')
    raise IndexError
```

## Definir x^2, ln(y) y x*ln(y)

```python
import math

x_pow2 = []
xlny = []
lny = []

for n in x_values:
    x_pow2.append(n**2)

for n in y_values:
    lny.append(math.log(n))

for n in range(len(x_values)):
    xlny.append(x_values[n]*lny[n])
```

## Realizar la suma de x, x^2 y ln(y)

```python
x_pow2_sum = 0
lny_sum = 0
xlny_sum = 0

for n in x_pow2:
    x_pow2_sum+=n

for n in lny:
    lny_sum+=n

for n in xlny:
    xlny_sum+=n
```

## Promedio de x y ln(y) 

```python
x_sum = 0
x_mean = 0
lny_sum = 0
lny_mean = 0

for n in x_values:
    x_sum+=n

for n in lny:
    lny_sum+=n

x_mean = x_sum/len(x_values)
lny_mean = lny_sum/len(lny)
```

## Hallar 1/x, 1/y, la multiplicación de ambas y 1/x^2

```python
x_inverse = []
y_inverse = []
xy_inverse = []
x_inverse_pow2 = []

for n in x_values:
    x_inverse.append(1/n)

for n in y_values:
    y_inverse.append(1/n)

for n in range(len(x_values)):
    xy_inverse.append(y_inverse[n]*x_inverse[n])

for n in x_inverse:
    x_inverse_pow2.append(n**2)
```

## Hallar el promedio de 1/x y de 1/y y sumatoria de 1/x, 1/y, la multiplicación de ambas y 1/x^2

```python
x_inverse_sum = 0
y_inverse_sum = 0
xy_inverse_sum = 0
x_inverse_pow2_sum = 0
x_inverse_mean = 0
y_inverse_mean = 0

for n in x_inverse:
    x_inverse_sum+=n

for n in y_inverse:
    y_inverse_sum+=n

for n in xy_inverse:
    xy_inverse_sum+=n

for n in x_inverse_pow2:
    x_inverse_pow2_sum+=n

x_inverse_mean = x_inverse_sum/len(x_values)
y_inverse_mean = y_inverse_sum/len(x_values)
```

## Hallar a0 y a1 para la ecuación de la linea

```python
a1_1 = (len(x_values)*xlny_sum-x_sum*lny_sum)/(len(x_values)*x_pow2_sum-(x_sum**2))

a0_1 = (lny_mean-a1*x_mean)
```

## Imprimir la linea hallada

```python
print(f'La ecuación de linea hallada es: y = {a0_1} + {a1_1}x')
```

## Se aplica la transformación para la función exponencial

```python
alpha_1 = math.exp(a0_1)
beta_1 = a1_1
```

## Imprimir la función hallada

```python
print(f'La ecuación hallada es: y = {alpha_1}e^{beta_1}x')
```

## Se aplica la transformación para la ecuación de razón de crecimiento

```python
a1_2 = (len(x_values)*xy_inverse_sum-x_inverse_sum*y_inverse_sum )/ (len(x_values)*x_inverse_pow2_sum-x_inverse_sum**2
)

a0_2 = y_inverse_mean - a1_2 * x_inverse_mean

alpha_2 = 1/a0_2
beta_2 = a1_2/a0_2
```

## Imprimir la función hallada

```python
print(f'La ecuación hallada es: y = {alpha_2}x/{beta_2}+X')
```

## Gráfica con la ecuación de la linea recta

```python
import matplotlib.pyplot as plt

plt.plot(x_values, y_values, 'o')
plt.plot(x_values, a0_1 + a1_1*x_values)
plt.grid()
```

## Gráfica con la función exponencia hallada

```python
plt.plot(x_values, y_values, 'o')
plt.plot(x_values,  alpha_1*np.exp(beta_1*x_values))
plt.grid()
```

## Gráfica con la ecuación de razón de crecimiento

```python
plt.plot(x_values, y_values, 'o')
plt.plot(x_values, alpha_2*(x_values/(beta_2+x_values)))
plt.grid()
```

# Regresión lineal por minimos cuadrados

## Definir valores de x y y

```python
import numpy as np

x_values = np.array([1,2,3,4,5])
y_values = np.array([5, 6.6, 8.5, 10.8, 14.2])

if(len(x_values)!=len(y_values)):
    print('La longitud de los puntos de x y y no es la misma.')
    raise IndexError
```

## Definir x^2, ln(y) y x*ln(y)

```python
import math

x_pow2 = []
xlny = []
lny = []

for n in x_values:
    x_pow2.append(n**2)

for n in y_values:
    lny.append(math.log(n))

for n in range(len(x_values)):
    xlny.append(x_values[n]*lny[n])
```

## Realizar la suma de x, x^2 y ln(y)

```python
x_pow2_sum = 0
lny_sum = 0
xlny_sum = 0

for n in x_pow2:
    x_pow2_sum+=n

for n in lny:
    lny_sum+=n

for n in xlny:
    xlny_sum+=n
```

## Promedio de x y ln(y) 

```python
x_sum = 0
x_mean = 0
lny_sum = 0
lny_mean = 0

for n in x_values:
    x_sum+=n

for n in lny:
    lny_sum+=n

x_mean = x_sum/len(x_values)
lny_mean = lny_sum/len(lny)
```

## Hallar a0 y a1

```python
a1 = (len(x_values)*xlny_sum-x_sum*lny_sum)/(len(x_values)*x_pow2_sum-(x_sum**2))

a0 = (lny_mean-a1*x_mean)
```

## Imprimir la linea hallada

```python
print(f'La ecuación de linea hallada es: y = {a0} + {a1}x')
```

## Se aplica la transformación inversa

```python
alpha = math.exp(a0)
beta = a1
```

## Imprimir la función hallada

```python
print(f'La ecuación hallada es: y = {alpha}e^{beta}x')
```

## Gráfica con la ecuación de la linea recta

```python
import matplotlib.pyplot as plt

plt.plot(x_values, y_values, 'o')
plt.plot(x_values, a0 + a1*x_values)
plt.grid()
```

## Gráfica con la función exponencia hallada

```python
plt.plot(x_values, y_values, 'o')
plt.plot(x_values,  alpha*np.exp(beta*x_values))
plt.grid()
```

## Gráfica con ambas funciones

```python
plt.plot(x_values, y_values, 'o')
plt.plot(x_values,  alpha*np.exp(beta*x_values))
plt.plot(x_values, a0 + a1*x_values)
plt.grid()
```
